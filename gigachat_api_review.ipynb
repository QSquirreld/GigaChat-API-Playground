# Необходимые библиотеки

```python
import getpass
import os
from gigachat import GigaChat
```

# Установка аутентификационных данных

```python
if "GIGACHAT_CREDENTIALS" not in os.environ:
    os.environ["GIGACHAT_CREDENTIALS"] = getpass.getpass("Введите ключ авторизации GigaChat API: ")
```

```python
# Создаём объект GigaChat, но НЕ выполняем запрос
giga = GigaChat(credentials=os.environ["GIGACHAT_CREDENTIALS"], verify_ssl_certs=False)
```

# Использование модели для генерации текста

```python
user_prompt = input("Введите ваш промпт: ")
```

# Генерируем ответ от модели

```python
response = giga.chat(user_prompt)
```

# Извлекаем текст из объекта ответа

```python
print(response.choices[0].message.content)
```

# Повторное использование модели

Для начала нужно создать переменную для хранения последнего обращения пользователя

```python
last_user_input = ""
```

Функция получения ответа модели

```python
def get_response(prompt):
    response = giga.chat(prompt)
    return response.choices[0].message.content
```

```python
while True:
    user_input = input("Вы: ")

    if user_input != last_user_input:

        model_response = get_response(user_input)

        print("Модель:", model_response)
```

# Потоковый вывод

Функция стриминга

```python
def stream_response(prompt, role="user"):
    print(f"\n{role.capitalize()}: {prompt}")
    print("Модель: ", end="", flush=True)

    for chunk in giga.stream(prompt):
        if chunk.choices[0].delta.content:
            print(chunk.choices[0].delta.content, end="", flush=True)  # Потоковый вывод
    print("\n")
```

```python
while True:
    user_input = input("\nВы: ")
    if user_input.lower() in ["exit", "пока"]:
        break  # Выход из цикла
    stream_response(user_input)
```
